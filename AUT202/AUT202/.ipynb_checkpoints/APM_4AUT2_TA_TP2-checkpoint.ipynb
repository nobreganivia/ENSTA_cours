## Ex.1: Vehicle Suspension System

Vehicle Suspension System



with the following system parameters:

    (m) mass 1.0 kg

    (k) spring constant 5.0 N/m

    $(c)$ damping constant 2 Ns/m

### **Equation of Motion:**

For a mass-spring-damper system, the equation of motion can be derived using Newton's second law or by applying the principles of dynamics. The general form of the equation is:

$$
m \frac{d^2x}{dt^2} + c \frac{dx}{dt} + kx = u + w
$$

### **State-Space Representation:**

To convert the second-order differential equation into a system of first-order differential equations, we introduce the following state variables:
- Let $ x_1 = x $ (position),
- Let $ x_2 = \dot x $ (velocity).

The system becomes:

$$\dot x = Ax+Bu+Dw$$

where
state vector $x = (position, velocity)$, and state and control matrices are the following:
$$ A = \begin{pmatrix} 0&1\\ -\frac{k}{m}&-\frac{c}{m}\end{pmatrix},\ B = \begin{pmatrix} 0\\ \frac{1}{m} 
\end{pmatrix},\ D =\begin{pmatrix} 0\\ \frac{1}{m} 
\end{pmatrix}$$

3. Let $x_0 = (0, 0).$ Let disturbance input $w(t)$ is given by a step function
    $$w = \left\{\begin{aligned}
        &0, \quad t <3.0\\
        &1.0,  \quad 3.0 \leq t < 7.0\\
        &0,  \quad t \geq 7

```python
import numpy as np
from scipy.integrate import odeint, solve_ivp
import matplotlib.pyplot as plt
import math

# x is a state vecot, t is a current time, A is a dynamic matrix, B is a control matrix, D is a disturbance matrix, 
# u_func(t) is a control input, w_func(t) is a disturbance input 
def StateSpace(x, t, A, B, D, u_func, w_func):
    w = w_func(t)
    u = u_func(t)
    return np.dot(A, x) + np.dot(B, u).flatten() + np.dot(D, w).flatten()

# OpenLoopInputOutputSystem with time-dependent w(t)
def OpenLoopInputOutputSystem(A, B, D, C, u_func, w_func, T, x0):
    # Solve the ODE system with odeint, passing the time-dependent w function
    solution = odeint(StateSpace, x0, T, args=(A, B, D, u_func, w_func))
    # Output calculation using C and the solution
    return (C @ solution.T) 

def BangBangController(A, B, D, C, u_const, ref, delay, w_func, T, x0):
    y_res = np.empty(x0.shape).reshape(-1,1)
    t0 = 0
    t1 = delay
    while t1 < T[-1]+delay:
        # Extract time segment
        T_bang = T[T<=t1]
        T = T[T_bang.shape[0]-1:]
        # Bang-Bang control logic
        u_func = lambda t: -u_const if C @ x0 - ref > 0 else (u_const if C @ x0 - ref < 0 else 0.0)
        # Compute system response
        sol = odeint(StateSpace, x0, T_bang, args=(A, B, D, u_func, w_func)).T
        x0 = sol[:,-1].flatten()
        # Update variables
        t0 = t1
        t1 += delay
        y_res = np.hstack((y_res, sol[:,1:]))
    return C @ y_res  
```

```python

# Paramètres du système
m = 1
k = 5
c = 2
g = 9.8
K_p = 400
K_i = 200
K_d = 50

A = np.array([[0, 1],
              [-k/m, -c/m]])  # Correction de c au lieu de rho

B = np.array([[0],
              [1/m]])  # Transformation en vecteur colonne

D = np.array([[0],
              [1/m]])  # Transformation en vecteur colonne

C = np.array([[1, 0]])  # Matrice de sortie

# État initial
x0 = np.array([0, 0])

# Paramètres de temps
t0 = 0  # Temps initial
tf = 15  # Temps final
T = np.linspace(t0, tf, 1000)  # Vecteur de temps

# Fonction d'entrée de perturbation
def w_func(t):
    if t < 3.0:
        return np.array([[0.0]])
    elif 3.0 <= t < 7.0:
        return np.array([[1.0]])
    else:
        return np.array([[0.0]])

# Fonction du contrôleur PID
def PIDController(K_p, K_d, K_i, A, B, D, C, u_const, x_ref, w_func, T, x0):
    def system_dynamics(t, x):
        # Calcul de l'erreur
        y = C @ x.reshape(-1, 1)  # S'assurer que y est un tableau 2D
        e = x_ref - y[0, 0]  # Erreur scalaire
        
        # Initialisation de l'intégrale et de la dérivée de l'erreur
        if not hasattr(system_dynamics, 'integral'):
            system_dynamics.integral = 0
        if not hasattr(system_dynamics, 'prev_error'):
            system_dynamics.prev_error = 0
        if not hasattr(system_dynamics, 'prev_time'):
            system_dynamics.prev_time = t0
        
        dt = max(t - system_dynamics.prev_time, 1e-6)  # Éviter dt = 0
        
        system_dynamics.integral = np.clip(system_dynamics.integral + e * dt, -1000, 1000)  # Éviter la saturation
        derivative = (e - system_dynamics.prev_error) / dt
        
        # Mise à jour des erreurs précédentes
        system_dynamics.prev_error = e
        system_dynamics.prev_time = t
        
        # Loi de commande PID
        u = np.clip(K_p * e + K_i * system_dynamics.integral + K_d * derivative, -100, 100)
        
        # Assurer que u est un vecteur colonne
        u = np.array([[u]])
        
        # Dynamique du système
        dxdt = A @ x.reshape(-1, 1) + B @ (u + u_const) + D @ w_func(t)
        return dxdt.flatten()
    
    # Initialisation des valeurs du contrôleur PID
    system_dynamics.integral = 0
    system_dynamics.prev_error = 0
    system_dynamics.prev_time = t0
    
    # Résolution du système avec solve_ivp
    sol = solve_ivp(system_dynamics, [t0, tf], x0, t_eval=T, method='RK45')  # Utilisation de RK45 pour plus de stabilité
    return sol.y

# Calcul de la réponse du système
u_const = np.array([[0.8]])
delay = 0.5
x_ref = 0.0
y_pid = PIDController(K_p, K_d, K_i, A, B, D, C, u_const, x_ref, w_func, T, x0)
y =  BangBangController(A, B, D, C, u_const, x_ref, delay, w_func, T, x0)



plt.plot(T, y[0,:], linewidth=2.0, color='red')
plt.plot(T, y_pid[0,:], linewidth=2.0, color='purple')

plt.grid(color='black', linestyle='--', linewidth=1.0, alpha=0.7)
plt.xlim([t0, tf])
plt.ylabel(r'Position (m)')
plt.xlabel('Time (s)')
plt.title('Active Suspension System ')
plt.plot(T, T* 0, linewidth=2.0, color='blue')
plt.legend(['bang-bang controller', 'PID controller','not controlled','reference input' ])
plt.show()    
# Tracé de la réponse
plt.plot(T, y_pid[0, :], linewidth=2.0, color='purple', label='PID Controller')
plt.plot(T, np.zeros_like(T), linewidth=2.0, color='blue', linestyle='--', label='Reference Input')
plt.grid(color='black', linestyle='--', linewidth=1.0, alpha=0.7)
plt.xlim([t0, tf])
plt.ylabel(r'Position (m)')
plt.xlabel('Time (s)')
plt.title('Système de suspension active - Suivi de référence PID')
plt.legend()
plt.show()
```

## DC Motor Speed: System Modeling

A common actuator in control systems is the DC motor. It directly provides rotary motion and, coupled with wheels or drums and cables, can provide translational motion. The electric equivalent circuit of the armature and the free-body diagram of the rotor are shown in the following figure.



For this example, we will assume that the input of the system is the voltage source ($V$) applied to the motor's armature, while the output is the rotational speed of the shaft $\dot{\theta}$. The rotor and shaft are assumed to be rigid. We further assume a viscous friction model, that is, the friction torque is proportional to shaft angular velocity. We will assume that the magnetic field is constant and, therefore, that the motor torque is proportional (with constant $K_t$) to only the armature current. Let us remark that in SI units the motor torque and back emf constants are equal, that is, $K_t = K_e$;

The physical parameters for our example are:

    (J)   moment of inertia of the rotor     0.01 kg.m^2

    (b)     motor viscous friction constant    0.1 N.m.s

    (Ke)    electromotive force constant       0.01 V/rad/sec

    (Kt)    motor torque constant              0.01 N.m/Amp

    (R)     electric resistance                1 Ohm

    (L)     electric inductance                0.5 H
    

Let us suppose that the measured output of the system is the angular velocity of the rotor $\dot{\theta}$ and the current intensity $\dot{i}$. Then the state space model of the system is the following:
$$\dot x = Ax + Bu$$

$$ y = Cx$$

where state vector $x=(\dot{\theta}, i)$, control vector $u=V$, and state and control matrices are the following

$$ A = \begin{pmatrix} -\frac{b}{J}&\frac{K}{J}\\ -\frac{K}{L}&-\frac{R}{L}\end{pmatrix},\ B = \begin{pmatrix} 0\\ \frac{1}{L} 
\end{pmatrix}$$

Let us suppose that we only interested in rotation speed (angular velocity of the rotor), i.e. $C=(1,0)$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.signal import lti, step
import math

# Motor DC parameters
J = 0.01  # Moment of inertia (kg.m^2)
b = 0.1   # Motor viscous friction constant (N.m.s)
Ke = 0.01  # Electromotive force constant (V/rad/sec)
Kt = 0.01  # Motor torque constant (N.m/Amp)
R = 1     # Electric resistance (Ohm)
L = 0.5   # Electric inductance (H)

# State-space representation
A = np.array([[-b/J, Kt/J],
              [-Ke/L, -R/L]])
B = np.array([[0],
              [1/L]])
C = np.array([[1, 0]])  # Measuring only angular velocity
D = np.array([[0]])

# Define PID controller parameters (Tuned using Ziegler-Nichols)
Kp = 120  # Proportional gain
Ki = 30  # Integral gain
Kd = 3 # Derivative gain

# Time settings
t0 = 0  # Initial time
tf = 5  # Final time
T = np.linspace(t0, tf, 1000)  # Time vector

# Initial state
x0 = np.array([0, 0]).flatten()

# Step input (1 rad/sec reference)
V_step = 1.0

def PIDController(Kp, Ki, Kd, A, B, C, T, x0, V_step):
    dt = T[1] - T[0]
    num_states = A.shape[0]
    x = np.zeros((num_states, len(T)))  
    x[:, 0] = x0.flatten()

    u = 0  
    integral = 0
    prev_error = 0

    y = np.zeros(len(T))  

    for i in range(1, len(T)):
        error = V_step - C @ x[:, i-1]

        integral += error * dt
        derivative = (error - prev_error) / dt
        u = Kp * error + Ki * integral + Kd * derivative
        prev_error = error

        dxdt = (A @ x[:, i-1].reshape(-1, 1) + B * u).flatten()
        x[:, i] = x[:, i-1] + dxdt.flatten() * dt
        y[i] = (C @ x[:, i]).item()

    return y
#PID compute
y_d1 = PIDController(Kp, Ki, 10, A, B, C, T, x0, V_step)
y_d2 = PIDController(Kp, Ki, 5, A, B, C, T, x0, V_step)
y_d3 = PIDController(Kp, Ki, 3, A, B, C, T, x0, V_step)

# Plot PID-controlled response
plt.figure()
plt.plot(T, y_d1, linewidth=2.0, color='purple')
plt.plot(T, y_d2, linewidth=2.0, color='green')
plt.plot(T, y_d3, linewidth=2.0, color='red')
plt.axhline(y=0.95, color='blue', linestyle='--')
plt.axhline(y=1.05, color='blue', linestyle='--')
plt.grid(color='black', linestyle='--', linewidth=1.0, alpha=0.7)
plt.xlim([t0, tf])
plt.ylabel('Angular Velocity (rad/sec)')
plt.xlabel('Time (s)')
plt.title('DC Motor Speed - PID Controller Tracking 1 rad/sec Reference')
plt.legend(['Kd = 10', 'Kd = 5', 'Kd = 3','Reference Input'])
plt.show()
```

## EX. 3 Inverted Pendulum

The system in this example consists of an inverted pendulum mounted to a motorized cart. The inverted pendulum system is an example commonly found in control system textbooks and research literature. Its popularity derives in part from the fact that it is unstable without control, that is, the pendulum will simply fall over if the cart isn't moved to balance it. Additionally, the dynamics of the system are nonlinear. The objective of the control system is to balance the inverted pendulum by applying a force to the cart that the pendulum is attached to. A real-world example that relates directly to this inverted pendulum system is the attitude control of a booster rocket at takeoff.



Let us consider the system with the following system parameters
    
    (M)       mass of the cart                         0.5 kg
    
    (m)       mass of the pendulum                     0.2 kg
    
    (l)       length to pendulum center of mass        0.3 m
    
    (b)       coefficient of friction for cart         0.1 N/m/sec
    
    (I)       mass moment of inertia of the pendulum   0.006 kg.m^2
    
    (F)       force applied to the cart
    
    (y)       cart position coordinate
    
    (theta)   angle between the pendulum and the vertical axis

$$ml\cos(\theta)\ddot{y} + (I+ml^2)\ddot{\theta} - mgl\sin\theta = 0$$

```python
import sympy as sp
# Answer to EX1 q1 using symbolic calculs
# Define symbolic variables
M, m, b, l, I, g, F = sp.symbols('M m b l I g F')
y, y1, theta, theta1, doty1, dottheta1 = sp.symbols('y y1 theta theta1 doty1 dottheta1')

# Define the differential equations of the system
eq1 = (M+m)*doty1 + b*y1 + m*l*dottheta1*sp.cos(theta) - m*l*theta1**2*sp.sin(theta) - F
eq2 = m*l*sp.cos(theta)*doty1 + (I+m*l**2)*dottheta1 - m*g*l*sp.sin(theta)

# Solve for the first derivative of theta1 (angular velocity)
dottheta1_sol = sp.solve(eq2, dottheta1)[0]

# Solve for the first derivative of y1 (linear velocity)
doty1_sol = sp.simplify(sp.solve(eq1.subs(dottheta1, dottheta1_sol), doty1)[0])
dottheta1_sol = sp.simplify(dottheta1_sol.subs(doty1,doty1_sol))

# Define the state-space representation of the system dynamics
f1 = y1
f2 = doty1_sol 
f3 = theta1
f4 = dottheta1_sol
f = sp.Matrix([f1, f2, f3, f4])

# Define state and control variables
variables_x = sp.Matrix([y,y1,theta,theta1])
variables_u = sp.Matrix([F])

# Compute the Jacobian matrices of the system
jacobian_A = sp.simplify(f.jacobian(variables_x).subs([(theta,0), (theta1,0)]))
jacobian_B = sp.simplify(f.jacobian(variables_u).subs([(theta,0), (theta1,0)]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define system parameters
M = 0.5     # Mass of the cart (kg)
m = 0.2     # Mass of the pendulum (kg)
l = 0.3     # Distance to the pendulum center of mass (m)
b = 0.1     # Coefficient of friction for the cart (N/m/s)
I = 0.006   # Moment of inertia of the pendulum (kg.m^2)
g = 9.81    # Acceleration due to gravity (m/s^2)
w_disturbance = 0.1  # Constant disturbance on the pendulum angle

# Compute auxiliary constants for the system dynamics
den = I * (M + m) + M * m * l**2  # Common denominator in the equations

# State-space representation matrices for the **linearized** inverted pendulum system
A = np.array([
    [0, 0, 1, 0],  # y_dot = dy
    [0, 0, 0, 1],  # theta_dot = dtheta
    [0, -g * m**2 * l**2 / den, -(I + m * l**2) * b / den, 0],  # Acceleration of the cart
    [0, g * m * l * (M + m) / den, m * l * b / den, 0]  # Angular acceleration of the pendulum
])

B = np.array([
    [0],  # Force has no direct effect on position
    [0],  # Force has no direct effect on angle
    [(I + m * l**2) / den],  # Force effect on cart acceleration
    [-m * l / den]  # Force effect on pendulum angular acceleration
])

C = np.array([
    [1, 0, 0, 0],  # Measuring cart position
    [0, 1, 0, 0]   # Measuring pendulum angle
])

W = np.array([
    [0],  # No disturbance directly affects position
    [0],  # No disturbance directly affects angle
    [0],  # No disturbance affects cart acceleration
    [1]   # Disturbance affects angular acceleration
])

# Initial state vector (cart at rest, small initial pendulum angle)
x0 = np.array([0, 0.05, 0, 0])  # Initial angle of 0.05 rad (small tilt)

# PID controller parameters for stabilizing the pendulum angle (theta)
Kp = 100   # Proportional gain
Ki = 50    # Integral gain
Kd = 10    # Derivative gain

# Simulation time settings
t0 = 0       # Start time (s)
tf = 5       # End time (s)
T = np.linspace(t0, tf, 1000)  # Time vector

# Initialize the integral error variable globally
integral_error = 0  

# Define the PID-controlled inverted pendulum system
def inverted_pendulum_pid(t, x, Kp, Ki, Kd):
    
    global integral_error  # Use global variable to store integral term
    
    y, theta, dy, dtheta = x  # Extract state variables

    # Compute the error (desired theta_ref = 0)
    error = 0 - theta  
    
    # Compute the integral of the error
    dt = T[1] - T[0]  # Time step
    integral_error += error * dt  # Numerical integration

    # Compute the derivative of the error
    derivative_error = -dtheta  

    # Compute the PID control force (control input u)
    u = Kp * error + Ki * integral_error + Kd * derivative_error

    # Compute the system dynamics: dx/dt = A*x + B*u + W*w
    dxdt = A @ x + B.flatten() * u + W.flatten() * w_disturbance
    
    return dxdt

# Solve the system using numerical integration
sol = solve_ivp(
    inverted_pendulum_pid, [t0, tf], x0, t_eval=T, args=(Kp, Ki, Kd)
)

# Extract simulated results
Y = sol.y[0]     # Cart position over time
Theta = sol.y[1]  # Pendulum angle over time

# Plot the results
plt.figure(figsize=(10, 5))

# Plot cart position
plt.subplot(2, 1, 1)
plt.plot(T, Y, 'b', linewidth=2)
plt.ylabel("Cart Position (m)")
plt.grid()

# Plot pendulum angle
plt.subplot(2, 1, 2)
plt.plot(T, Theta, 'r', linewidth=2)
plt.axhline(y=0, color='black', linestyle='--', label="Reference")
plt.ylabel("Pendulum Angle (rad)")
plt.xlabel("Time (s)")
plt.grid()
plt.legend()

plt.show()
```
